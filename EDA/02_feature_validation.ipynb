# 02. Feature Engineering 검증

## 목표
- 피처 생성 과정 검증
- 컬럼명 확인 및 데이터 타입 체크
- 생성된 피처 개수 확인
- 결측치 및 상수 컬럼 분석

```python
import sys
sys.path.append('../utils')
sys.path.append('../features')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from common_utils import load_data
from feature_engineering import build_a_features, build_b_features
from preprocessing import FeaturePreprocessor

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

print("✅ 라이브러리 로드 완료")
```

## 1. 원본 데이터 컬럼 확인

```python
train, test, train_a, train_b, test_a, test_b = load_data()

print("=" * 60)
print("Train A 데이터")
print("=" * 60)
print(f"Shape: {train_a.shape}")
print(f"\n컬럼 목록:")
print(train_a.columns.tolist())
print(f"\n데이터 타입:")
print(train_a.dtypes)
print(f"\n샘플 데이터 (첫 3행):")
print(train_a.head(3))
```

```python
print("=" * 60)
print("Train B 데이터")
print("=" * 60)
print(f"Shape: {train_b.shape}")
print(f"\n컬럼 목록:")
print(train_b.columns.tolist())
print(f"\n데이터 타입:")
print(train_b.dtypes)
print(f"\n샘플 데이터 (첫 3행):")
print(train_b.head(3))
```

## 2. Feature Engineering 실행

```python

features_a = build_a_features(train_a, stats=None, label_df=None)

print("\n" + "=" * 60)
print(f"생성된 A 피처 Shape: {features_a.shape}")
print(f"컬럼 수 (Test_id 제외): {features_a.shape[1] - 1}개")
print(f"\n컬럼 목록 (앞 20개):")
print(features_a.columns.tolist()[:20])
```

```python

features_b = build_b_features(train_b, label_df=None)

print(f"생성된 B 피처 Shape: {features_b.shape}")
print(f"컬럼 수 (Test_id 제외): {features_b.shape[1] - 1}개")
print(f"\n컬럼 목록 (앞 20개):")
print(features_b.columns.tolist()[:20])
```

## 3. 병합 및 결측치 분석

```python
# 병합 함수
train_features = pd.concat([
    train[train['Test'] == 'A'].merge(features_a, on='Test_id', how='left'),
    train[train['Test'] == 'B'].merge(features_b, on='Test_id', how='left')
], ignore_index=True)

X = train_features.drop(['Test_id', 'Test', 'Label'], axis=1, errors='ignore')
y = train_features['Label']

# 숫자형 컬럼만 선택
X = X.select_dtypes(include=[np.number])

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"총 피처 개수: {X.shape[1]}개")
print(f"위험군 비율: {y.mean():.2%}")
```

```python
# 결측치 분석
missing = X.isnull().sum()
missing_pct = (missing / len(X)) * 100
missing_df = pd.DataFrame({
    'column': X.columns,
    'missing_count': missing.values,
    'missing_pct': missing_pct.values
}).sort_values('missing_pct', ascending=False)

print("\n결측치 상위 10개 컬럼:")
print(missing_df.head(10))

print(f"\n전체 결측치 비율: {X.isnull().sum().sum() / (X.shape[0] * X.shape[1]) * 100:.2f}%")
```

```python
# 상수 컬럼 체크
constant_cols = []
low_variance_cols = []

for col in X.columns:
    unique_count = X[col].nunique(dropna=True)
    valid_count = X[col].notna().sum()
    
    if unique_count <= 1:
        constant_cols.append((col, unique_count, valid_count))
    elif unique_count < 5:
        low_variance_cols.append((col, unique_count, valid_count))

print("\n" + "=" * 60)
print("상수/저분산 컬럼 분석")
print("=" * 60)
print(f"상수 컬럼 (고유값 <= 1): {len(constant_cols)}개")
if constant_cols:
    print("\n상수 컬럼 목록:")
    for col, unique, valid in constant_cols[:10]:
        print(f"  {col}: 고유값 {unique}개, 유효값 {valid}개")

print(f"\n저분산 컬럼 (고유값 < 5): {len(low_variance_cols)}개")
if low_variance_cols:
    print("\n저분산 컬럼 예시:")
    for col, unique, valid in low_variance_cols[:5]:
        print(f"  {col}: 고유값 {unique}개, 유효값 {valid}개")
```

## 4. 전처리 적용 및 비교

```python
preprocessor = FeaturePreprocessor()
X_processed = preprocessor.fit_transform(X)

print(f"원본 피처 개수: {X.shape[1]}개")
print(f"전처리 후 피처 개수: {X_processed.shape[1]}개")
print(f"제거된 피처 개수: {X.shape[1] - X_processed.shape[1]}개")
print(f"\n전처리 후 결측치: {X_processed.isnull().sum().sum()}개")
```

## 5. 피처 분포 시각화

```python
# 주요 피처 분포 (상위 9개)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for idx, col in enumerate(X_processed.columns[:9]):
    axes[idx].hist(X_processed[col].dropna(), bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    axes[idx].set_title(col, fontsize=10)
    axes[idx].set_xlabel('Value')
    axes[idx].set_ylabel('Frequency')
    axes[idx].grid(alpha=0.3)

plt.suptitle('주요 피처 분포 (상위 9개)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../output/figures/feature_distributions.png', dpi=150, bbox_inches='tight')
plt.show()

print("✅ 피처 분포 저장: ../output/figures/feature_distributions.png")
```
